**PROGRAMA 5:** Tienda de juguetes

```
def tienda (n_p,n_jm,pedidos):
  peso_p=n_p*150
  peso_jm=n_jm*170
  peso_tot=(peso_p+peso_jm)/1000
  return peso_tot

def extra(gramos_10kg):
   g_extra=(tienda (n_p,n_jm,pedidos)*1000)-gramos_10kg
   print("ERROR: el pedido sobrepasa a los 10kg por",g_extra,"gramos extra")
   return False
    

num_pedidos=0
gramos_10kg=10000
pedidos=int (input("Ingresa el número de pedidos:"))
while (num_pedidos<pedidos):
  num_pedidos+=1
  print("\nPedido número",num_pedidos)
  n_p=int (input("Cantidad de payasos:"))
  n_jm=int (input("Cantidad de juegos de mesa:"))
  print("El peso del pedido",num_pedidos,"es:",tienda (n_p,n_jm,pedidos),"kilogramos")
  if (tienda (n_p,n_jm,pedidos)>10):
    print(extra(gramos_10kg))
```

**PROGRAMA 6:** Impuestos

```
def impuesto(salario):
 while (0>=salario):
   salario=int (input("El salario debe de ser mayor a 0, ingresa un nuevo salario:"))
 if (salario<=10000):
  print("El salario se encuentra en el rango de 0-10,000 pesos tiene descuento de impuesto del 5%")
  imp5=salario*0.05
  desc=salario-imp5
  return desc, salario<=10000
 elif (10000<salario<=15000):
  print("El salario se encuentra en el rango de 10,000-15,000 pesos tiene descuento de impuesto del 10%")
  imp10=salario*0.1
  desc=salario-imp10
  return desc, 10000<salario<=15000
 elif (15000<salario<=20000):
  print("El salario se encuentra en el rango de 15,000-20,000 pesos, tiene descuento de impuesto del 15%")
  imp15=salario*0.15
  desc=salario-imp15
  return desc, 15000<salario<=20000
 elif (20000<salario):
  print("El salario se encuentra en el rango de 20,000 pesos en adelante, tiene descuento de impuesto del 18%")
  imp18=salario*0.18
  desc=salario-imp18
  return desc, 20000<salario

salario=float (input("Ingresa el salario mensual del trabajador:"))
print("El salario con el descuento es: ",impuesto(salario))
```

**PROGRAMA 9:** Factorial

```
def factorial(num,constante):
  if (num>0):
   while (1<constante):
     constante-=1
     num*=constante
     fact=num
   return fact 
  elif (num<0):
    while (-1>constante):
     constante+=1
     num*=constante
     fact_neg=num
    return fact_neg

num=int (input("Ingresa un numero: "))
constante=num
print("El factorial del numero ingresado es:",factorial(num,constante))
```

**PROGRAMA 10:** Área de un círculo y volumen de un cilindro

```
def area(radio):
 A=3.1416*(radio**2)
 return A
def volumen(radio,h):
 V=area(radio)*h
 return V

radio=float (input("Ingresa el radio del círculo: "))
h=float (input("Ingresa la altura del cilindro: "))
print("El área del círculo es:",area(radio))
print("El volumen del cilindro es:",volumen(radio,h))
```
